```python
from scipy import stats
# Get numeric columns only
numeric_columns = sdf.select_dtypes(include=[np.number])

# Calculate Coefficient of Variation (CV)
cv_values = (numeric_columns.std() / numeric_columns.mean()) * 100

# Calculate 95% Confidence Interval (CI)
confidence_intervals = {}
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

for col in numeric_columns:
    mean = numeric_columns[col].mean()
    std_dev = numeric_columns[col].std()
    n = numeric_columns[col].count()  # Sample size
    margin_of_error = z_score * (std_dev / np.sqrt(n))
    
    confidence_interval = (mean - margin_of_error, mean + margin_of_error)
    confidence_intervals[col] = confidence_interval

# Display results
print("Coefficient of Variation (CV) for each column:\n", cv_values)
print("\n95% Confidence Interval (CI) for each column:\n", confidence_intervals)
```

```python
import matplotlib.pyplot as plt

# Plot the Confidence Intervals
plt.figure(figsize=(10, 6))
for col in numeric_columns.columns:
    plt.errorbar(x=col, y=numeric_columns[col].mean(), 
                 yerr=z_score * (numeric_columns[col].std() / np.sqrt(numeric_columns[col].count())), 
                 fmt='o', label=col)

plt.title('95% Confidence Intervals for Numeric Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select categorical columns from sdf
categorical_columns = sdf.select_dtypes(include=['object']).columns

# Descriptive statistics for each categorical column
for col in categorical_columns:
    print(f"\n--- {col} ---")
    
    # Count and Frequency
    counts = sdf[col].value_counts()
    print("Count:\n", counts)
    
    # Proportion
    proportions = sdf[col].value_counts(normalize=True)
    print("\nProportion:\n", proportions)
    
    # Mode
    mode_value = sdf[col].mode()[0]
    print("\nMode:", mode_value)
    
    # Minimum and Maximum (for categorical, this means the first and last alphabetically or based on order)
    min_value = sdf[col].min()
    max_value = sdf[col].max()
    print(f"\nMinimum: {min_value}")
    print(f"Maximum: {max_value}")

    # Rank (based on frequency)
    ranked = counts.rank(ascending=False)
    print("\nRank:\n", ranked)

# Visualization: Frequency bar plots for each categorical column
for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=sdf, x=col, palette='Set2')
    plt.title(f"Frequency Distribution of {col}")
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Correlation between categorical and numerical data (if any)
# Convert categorical data into numerical using pd.get_dummies for correlation analysis
categorical_dummies = pd.get_dummies(sdf[categorical_columns], drop_first=True)
correlation_matrix = pd.concat([sdf.select_dtypes(include=[np.number]), categorical_dummies], axis=1).corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix between Numerical and Categorical Data')
plt.show()

# Cramer's V for categorical correlation
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

# Example Cramer's V calculation for two categorical variables
for i, col1 in enumerate(categorical_columns):
    for col2 in categorical_columns[i+1:]:
        confusion_matrix = pd.crosstab(sdf[col1], sdf[col2])
        cramer_v_value = cramers_v(confusion_matrix)
        print(f"Cramer's V between {col1} and {col2}: {cramer_v_value:.4f}")

        # Heatmap for the relationship between col1 and col2
        plt.figure(figsize=(10, 6))
        sns.heatmap(confusion_matrix, annot=True, cmap='YlGnBu', fmt='g')
        plt.title(f"Heatmap of '{col1}' vs '{col2}'")
        plt.show()
```

# Inferential Analysis of Non-Categorical Data

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
import statsmodels.api as sm
# Assuming 'Value' and 'Quantity' are numeric columns
numeric_columns = ['Quantity', 'Value', 'Weight']

# Test of Mean: t-test
def t_test(data, column, popmean=0):
    t_stat, p_value = stats.ttest_1samp(data[column], popmean)
    return t_stat, p_value

# Test of Variance: Levene's Test
def levene_test(data, column):
    w_stat, p_value = stats.levene(*[group[column].dropna() for name, group in data.groupby('Category')])
    return w_stat, p_value

# Test of Proportion: Chi-square using scipy
def chi_square_test(data, column1, column2):
    contingency_table = pd.crosstab(data[column1], data[column2])
    chi2, p, dof, ex = stats.chi2_contingency(contingency_table)
    return chi2, p

# Test of Correlation: Pearson Correlation
def correlation_test(data, column1, column2):
    correlation, p_value = stats.pearsonr(data[column1], data[column2])
    return correlation, p_value

# Test of Normality: Shapiro-Wilk Test
def normality_test(data, column):
    stat, p_value = stats.shapiro(data[column].dropna())
    return stat, p_value

# Non-Parametric Test: Mann-Whitney U test
def mann_whitney_test(data, column, group1, group2):
    group1_data = data[data['Category'] == group1][column]
    group2_data = data[data['Category'] == group2][column]
    u_stat, p_value = stats.mannwhitneyu(group1_data, group2_data)
    return u_stat, p_value

# Test of Variance: Bartlett's Test
def bartlett_test(data, column):
    groups = [group[column].dropna() for name, group in data.groupby('Category')]
    stat, p_value = stats.bartlett(*groups)
    return stat, p_value

# Example usage
t_stat, t_p = t_test(sdf, 'Quantity', popmean=1000)
print(f"T-test: t-stat={t_stat}, p-value={t_p}")

w_stat, w_p = levene_test(sdf, 'Value')
print(f"Levene's Test: W-stat={w_stat}, p-value={w_p}")

chi2_stat, chi2_p = chi_square_test(sdf, 'Category', 'Shipping_Method')
print(f"Chi-square Test: chi2-stat={chi2_stat}, p-value={chi2_p}")

corr_stat, corr_p = correlation_test(sdf, 'Quantity', 'Value')
print(f"Correlation Test: stat={corr_stat}, p-value={corr_p}")

shapiro_stat, shapiro_p = normality_test(sdf, 'Weight')
print(f"Shapiro-Wilk Test: stat={shapiro_stat}, p-value={shapiro_p}")

mann_whitney_u, mann_whitney_p = mann_whitney_test(sdf, 'Value', 'Furniture', 'Electronics')
print(f"Mann-Whitney Test: U-stat={mann_whitney_u}, p-value={mann_whitney_p}")

bartlett_stat, bartlett_p = bartlett_test(sdf, 'Quantity')
print(f"Bartlett's Test: stat={bartlett_stat}, p-value={bartlett_p}")
```
